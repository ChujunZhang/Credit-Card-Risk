```python
def mul_histogram(nrow,ncol,nbins,tablename,colname):
    plt.figure(figsize=(16,16))
    for i,varname in enumerate(colname):
        plt.subplot(nrow,ncol,i+1)
        tablename[varname].hist(bins=nbins)
        plt.title(varname)
    return plt
```

```python
mul_histogram(2,3,20,bill,bill.columns)
plt.show()
```

```python
mul_histogram(2,3,20,pay,pay.columns)
plt.show()
```

# Data Cleaning

```python
data.isnull().sum() #no missing value
```

```python
#in education, since 0.5.6 not documented, so put them in others (4)
#in marriage,0 not documented, put in others(3)
fil=(data.EDUCATION==5) | (data.EDUCATION==6) | (data.EDUCATION==0)
data.loc[fil,'EDUCATION']=4
data.loc[data.MARRIAGE==0,'MARRIAGE']=3
```

```python
#for pay status
#-2 means no activity this month
#-1 is pay all
#0 means pay the minimum
#so set -1,-2 as 0
def fil_column(table,columnname,replace): #-1,-2 replace as 0
    for i in columnname:
        fil=(table[i]==-1)|(table[i]==-2)
        table.loc[fil,i]=replace
        
col=['PAY_1','PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
fil_column(data,col,0)
paystatus=data[['PAY_1','PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']]
mul_histogram(2,3,10,paystatus,paystatus.columns)
plt.show()
```
